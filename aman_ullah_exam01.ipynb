operation. (5 Marks)

- `Slice` the sorted list to obtain a sublist containing the middle three elements. (3 Marks)

### **Answer**

```python
nums = [3, 6, 2, 7, 5, 6, 8, 5, 8, 3, 7]
```

```python
# removing dupicates
def removing_duplicate(nums):
  num = []
  for i in nums:
    if i  in num:
      continue
    else:
      num.append(i)
  return num



```
