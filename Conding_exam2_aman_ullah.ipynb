```python
_input = input('enter the sentence')
def count_word(snc):
  vowels= 'aeiouAEIOU'
  words = len(snc.split())
  cherecters = len(snc)
  vowel_counts = sum(snc.count(vowel) for vowel in vowels)

  print('Total word number is :',words)
  print('Total charecters are : ',cherecters)
  print('Total vowel numbers :', vowel_counts)

count_word(_input)
```

```python

def factorials(num):
  factorial =1
  try:
    if num>0:
      for i in range(1,num+1):
        factorial *=i
    else :
      print('inter positive num')
  except :
    print('Erore : number is not valid')
  print(f"The factorial of  {num} is {factorial}")
```

```python
factorials(8)
```

```python
factorials(-8)
```

## Answer to the question no 02

**A**

```python
numbers = [3, 7, 2, 9, 12, 15, 8, 6]
new_list = []
for i in numbers:

  new_list.append(i**2)
print(new_list)
```

```python
for i in range(0, len(new_list)):
    for j in range(i+1, len(new_list)):
        if new_list[i] >= new_list[j]:
            new_list[i], new_list[j] = new_list[j],new_list[i]

# sorted list
print("Sorted List", new_list)
```
